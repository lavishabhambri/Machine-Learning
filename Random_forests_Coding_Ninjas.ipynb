# Data Analysis

```python
del df1['Name']
del df1['Ticket']
del df1['Fare']
del df1['Cabin']

del df2['Name']
del df2['Ticket']
del df2['Fare']
del df2['Cabin']
```

```python
df1
```

```python
df2
```

```python
#on Sex column
def getGender(s):
    if s =='male':
        return 0
    else:
        return 1
df1['Sex'] = df1.Sex.apply(getGender)
print(df1)

df2['Sex'] = df2.Sex.apply(getGender)
print(df2)
```

```python
#on embarked column
def Embarked(s):
    if s=='C':
        return 0
    elif s=='Q':
        return 1
    else:
        return 2
    
df1['Embarked'] = df1.Embarked.apply(Embarked)
print(df1)

df2['Embarked'] = df2.Embarked.apply(Embarked)
print(df2)
```

```python
df1.isnull().sum()
```

```python
#filling the nan values of age column of df1 by the mean of the persons survived and unsurvived respectively.

#suvived
s = df1[df1.Survived == 1].Age.mean()
df1["Age"]=np.where(pd.isnull(df1.Age) & df1["Survived"]==1  ,s, df1["Age"])

us=df1[df1.Survived == 0].Age.mean()
df1.Age.fillna(us,inplace=True)
df1
```

# Apply model

```python
train_data = df1.values
```

```python
x = train_data[:,0:6]
y = train_data[:,6:7]
```

```python
from sklearn import model_selection
x_train, x_test, y_train, y_test = model_selection.train_test_split(x, y)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(random_state = 0, max_depth = 10)
clf.fit(x_train, y_train)
print(clf.score(x_train, y_train))
print(clf.score(x_test, y_test))
```

# Decision trees on titanic dataset

```python
from sklearn.tree import DecisionTreeClassifier
clf1 = DecisionTreeClassifier()
clf1.fit(x_train, y_train)
print(clf1.score(x_train, y_train))
print(clf1.score(x_test, y_test))
```
