# Add Features to Boston Dataset

For the Boston dataset, your task is to add all 2 degree features to the dataset and compare scores for the original dataset and new dataset with added features.

```python
from sklearn import datasets
boston = datasets.load_boston()
```

```python
boston
```

```python
x = boston.data
y = boston.target
```

```python
print(x.shape)
print(y.shape)
```

```python
import pandas as pd
df = pd.DataFrame(x)
df
```

```python
df.columns = boston.feature_names
df
```

```python
boston.feature_names
```

# Adding features in dataset

```python
for i in range(13):
    for j in range(13):
        df[boston.feature_names[i]+'_'+ boston.feature_names[j]] = df[boston.feature_names[i]] * df[boston.feature_names[j]]
    
```

```python
df
```

```python
x2 = df.values
```

```python
from sklearn import model_selection
x1_train, x1_test, y1_train, y1_test = model_selection.train_test_split(x, y, random_state = 0)
x2_train, x2_test, y2_train, y2_test = model_selection.train_test_split(x2, y, random_state = 0)
```

# Get Classifier

```python
from sklearn.linear_model import LinearRegression
lr1 = LinearRegression()
lr2 = LinearRegression()
```

# Training the model

```python
lr1.fit(x1_train, y1_train)
lr2.fit(x2_train, y2_train)
```

# Predictions

```python
y1_pred = lr1.predict(x1_test)
y2_pred = lr2.predict(x2_test)
```

# Scores Comparison

```python
print('Linear regression : ', lr1.score(x1_test, y1_test))
print('Multivariate Linear regression : ', lr2.score(x2_test, y2_test))
```

### CONCLUSION :: So the score has decreased when we added so many features to the model. It becomes the case of OVERFITTING.
